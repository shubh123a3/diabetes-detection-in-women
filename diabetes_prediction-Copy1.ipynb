```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats


# Extract the column
data = df['Pregnancies']

# Create the figure and axes
fig, axes = plt.subplots(1, 4, figsize=(24, 6))

# KDE plot
sns.kdeplot(data, ax=axes[0])
axes[0].set_title('KDE Plot')

# Boxplot
sns.boxplot(data, ax=axes[1])
axes[1].set_title('Boxplot')

# QQ plot
stats.probplot(data, dist="norm", plot=axes[2])
axes[2].get_lines()[1].set_color('r')
axes[2].set_title('QQ Plot')

# Distplot
sns.histplot(data, kde=True, ax=axes[3])
axes[3].set_title('Distplot')

plt.tight_layout()
plt.show()
```

```python
new_df_cap = df.copy()

new_df_cap['Pregnancies'] = np.where(
    new_df_cap['Pregnancies'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Pregnancies'] < lower_limit,
        lower_limit,
        new_df_cap['Pregnancies']
    )
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
x=new_df_cap.drop(columns=['Outcome'])
y=new_df_cap['Outcome']
X_train, X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
new_df_cap.sample()
```

```python
X_train.sample()
```

```python
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
trf1=ColumnTransformer([('standar_scaler_all',StandardScaler(),[0,1,2,3])],remainder='passthrough')
```

```python
trf2=LogisticRegression()
```

```python
pipe = make_pipeline(trf1,trf2)
```

```python
pipe.fit(X_train,y_train)
```

```python
prediction=pipe.predict(X_test)
```

```python
score=pipe.score(X_test,y_test)
score
```
